```python
import gymnasium as gym
from gymnasium.envs.registration import register

import numpy as np
import random
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
try:
    register(
        id="SpaceInvaders-v0",
        entry_point="ale_py.env:AtariEnv",
        kwargs={"game": "space_invaders"},
        max_episode_steps=108000,
    )
    print("SpaceInvaders-v0 registrado correctamente")
except:
    print("SpaceInvaders-v0 ya estaba registrado")
```

```python
env = gym.make("SpaceInvaders-v0")
print("Número de acciones:", env.action_space.n)
env.close()
```

```python
def preprocess(obs):
    # Convertir a grayscale
    obs = np.mean(obs, axis=2).astype(np.uint8)
    # Redimensionar
    obs = obs[::2, ::2]
    return obs
```

```python
env = gym.make("SpaceInvaders-v0", render_mode="rgb_array")
obs, _ = env.reset()
obs = preprocess(obs)

plt.imshow(obs, cmap="gray")
plt.title("Frame preprocesado")
plt.axis("off")
plt.show()

env.close()
```

```python
from collections import deque

class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)

        return (
            torch.tensor(np.array(states), dtype=torch.float32),
            torch.tensor(actions, dtype=torch.int64),
            torch.tensor(rewards, dtype=torch.float32),
            torch.tensor(np.array(next_states), dtype=torch.float32),
            torch.tensor(dones, dtype=torch.float32),  # 👈 CLAVE
        )

    def __len__(self):
        return len(self.buffer)
```

```python
class DQN(nn.Module):
    def __init__(self, n_actions):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(105 * 80, 512),
            nn.ReLU(),
            nn.Linear(512, n_actions)
        )

    def forward(self, x):
        return self.net(x)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

policy_net = DQN(env.action_space.n).to(device)
target_net = DQN(env.action_space.n).to(device)
target_net.load_state_dict(policy_net.state_dict())

optimizer = optim.Adam(policy_net.parameters(), lr=1e-4)
buffer = ReplayBuffer(100_000)
```

```python
BATCH_SIZE = 32
GAMMA = 0.99
EPSILON_START = 1.0
EPSILON_END = 0.1
EPSILON_DECAY = 1_000_000
TARGET_UPDATE = 10_000
```

```python
def clip_reward(reward):
    return np.sign(reward)
```

```python
env = gym.make("SpaceInvaders-v0", render_mode="rgb_array")
state, _ = env.reset()
state = preprocess(state)

epsilon = EPSILON_START
step_count = 0
episode_reward = 0
rewards_history = []

for episode in range(200):  # puedes aumentar luego
    done = False
    episode_reward = 0

    while not done:
        step_count += 1
        epsilon = max(EPSILON_END, EPSILON_START - step_count / EPSILON_DECAY)

        if random.random() < epsilon:
            action = env.action_space.sample()
        else:
            with torch.no_grad():
                s = torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(device)
                action = policy_net(s).argmax().item()

        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        reward = clip_reward(reward)   # 🔥 REWARD CLIPPING 🔥
        next_state = preprocess(next_state)

        buffer.push(state, action, reward, next_state, done)
        state = next_state
        episode_reward += reward

    if len(buffer) >= BATCH_SIZE:
        states, actions, rewards, next_states, dones = buffer.sample(BATCH_SIZE)

        states = states.to(device).float()
        next_states = next_states.to(device).float()
        actions = actions.to(device).long()
        rewards = rewards.to(device).float()
        dones = dones.to(device).float()

        q_values = policy_net(states).gather(1, actions.unsqueeze(1)).squeeze()
        next_q = target_net(next_states).max(1)[0]
        target = rewards + GAMMA * next_q * (1 - dones)

        loss = nn.MSELoss()(q_values, target.detach())

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()


        if step_count % TARGET_UPDATE == 0:
            target_net.load_state_dict(policy_net.state_dict())

    rewards_history.append(episode_reward)
    print(f"Episodio {episode} | Reward: {episode_reward}")

env.close()
```

```python
plt.figure(figsize=(10,5))
plt.plot(rewards_history, label="Reward por episodio")

# media móvil
window = 10
if len(rewards_history) >= window:
    moving_avg = [
        np.mean(rewards_history[max(0, i-window):i+1])
        for i in range(len(rewards_history))
    ]
    plt.plot(moving_avg, label="Media móvil (10)", linewidth=2)

plt.xlabel("Episodio")
plt.ylabel("Reward (clipped)")
plt.title("Entrenamiento DQN - SpaceInvaders-v0")
plt.legend()
plt.grid()
plt.show()
```

```python
def test_agent(env, model, episodes=10):
    model.eval()
    rewards = []

    for ep in range(episodes):
        state, _ = env.reset()
        state = preprocess(state)
        done = False
        total_reward = 0

        while not done:
            with torch.no_grad():
                s = torch.tensor(state, dtype=torch.float32).unsqueeze(0).to(device)
                action = model(s).argmax().item()

            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            state = preprocess(next_state)
            total_reward += reward

        rewards.append(total_reward)
        print(f"Test episodio {ep}: {total_reward}")

    print("Media recompensa test:", sum(rewards)/len(rewards))
```
